```python
import pandas as pd
from shapely.geometry import Point

df = pd.read_excel('City.xlsx')
```

```python
df.head()
```

```python
# Fonction pour transformer la longitude et latitude en WKT (Well-Known Text)
def lat_lng_to_wkt(lat, lng):
    point = Point(lng, lat)  # Créer un point avec la longitude et la latitude
    return point.wkt  # Convertir le point en format WKT

# Appliquer la fonction à chaque ligne du DataFrame
df['WKT'] = df.apply(lambda row: lat_lng_to_wkt(row['lat'], row['lng']), axis=1)

# Sauvegarder les données avec la nouvelle colonne WKT dans un nouveau fichier CSV
df.to_csv('City_WKT.csv', index=False)
```
